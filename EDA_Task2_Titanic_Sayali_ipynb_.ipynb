```python
import pandas as pd

df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
print("Shape:",df.shape)
#Data Types and non null values
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,10),color='lightblue',edgecolor='black')
plt.suptitle("Histogram of Numeric Features",fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# prompt: boxplot for age

import seaborn as sns

sns.boxplot(x=df['Age'])
plt.title("Boxplot - Age")
plt.show()
```

```python
numeric_df=df.select_dtypes(include='number')

#correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(),annot=True,
cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

```python
df.skew(numeric_only=True)
```

## Final Observation
- 'Age','Cabin',and 'Embarked' columns have missing values.
- 'Fare' is Highly Skewed,as shown in the Histogram and Skewness Check.
- 'Pclass' and 'Sex' are Important Features affecting Survival.
- Boxplot revealed outliers in 'Age' and 'Fare'.
- Correltion heatmap shows weak to moderate correlation between numeric features and survival.
